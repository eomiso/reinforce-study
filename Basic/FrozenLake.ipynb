```python
# plot the game scores of frozenlake random agent

import numpy as np
import matplotlib.pyplot as plt
import gym

env = gym.make('FrozenLake-v0')



n_games = 1000
win_pct = []
scores = []
for i in range(n_games):
    done = False
    obs = env.reset()
    score = 0
    while not done:
        action = env.action_space.sample()
        obs, rew, done, info = env.step(action)
        
        score = rew
    scores.append(score)
    
    if i % 10 == 0:
        average = np.mean(scores[-10:])
        win_pct.append(average)

print(len(win_pct))        
plt.plot(win_pct)
plt.show()
```
